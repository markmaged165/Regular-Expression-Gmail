## 2- Extract the name [First + second] and add it as column

```python
name=[]
for row in df['Messy Text'] :

    rex=r'[A-Z]+[a-z]+\s[A-Z]+[a-z]+'
    # if re.match(rex,row)==None:
    #     name.append(None)
        # continue
    x=re.findall(rex,row)
    name.append(x[0])
    
print(name)
```

## 4- Extract the Phone Number and add it as column

```python
for i in range (0,5):
    print(df['Messy Text'][i])
```

```python
ph_numper=[]
for row in df['Messy Text'] :
    rex=r"\d{11}"
    match = re.search(rex ,row)
    if match:
        x=re.findall(rex,row)
        ph_numper.append(x[0])
    else:
        ph_numper.append(None)    

print(ph_numper)
```
